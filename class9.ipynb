```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris['Species'] = iris['Species'].replace(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], [0, 1, 2])
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'], c=iris['Species'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'], c=iris['Species'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.colorbar(label='Species')
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap = 'jet', alpha = 0.7)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.colorbar(label='Species')
plt.show()
```

```python
# annotation
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

plt.scatter(x, y)
plt.text(1, 10, 'Point 1')
plt.text(2, 20, 'Point 2')
plt.text(3, 25, 'Point 3')
plt.text(4, 30, 'Point 4')
plt.text(5, 35, 'Point 5', fontdict={'color': 'red', 'size': 12})
plt.show()
```

```python
# bar plot

x = ['A', 'B', 'C', 'D']
y = [3, 7, 2, 5]

plt.bar(x, y)
```

```python
plt.figure(figsize=(5, 2))
plt.bar(x, y, color='green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
# bar chart  with multiple color usinf list

x = ['A', 'B', 'C', 'D']
y = [3, 7, 2, 5]
colors = ['red', 'green', 'blue', 'yellow']

plt.bar(x, y, color=colors)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

```python
branch = ['CSE', 'ECE', 'IT', 'MECH']
students = [45, 35, 25, 30]
plt.bar(branch, students, alpha=0.6, width=0.5, edgecolor='black', linewidth=2, color = ['red', 'green', 'blue', 'yellow'])
plt.xlabel('Branch')
plt.ylabel('No of Students')
plt.title('Branch vs Students')
plt.show()
```

```python
branch = ['CSE', 'ECE', 'IT', 'MECH']
students = [45, 35, 25, 30]
plt.barh(branch, students, alpha=0.6, height=0.5, edgecolor='black', linewidth=2, color = ['red', 'green', 'blue', 'yellow'])
plt.xlabel('Branch')
plt.ylabel('No of Students')
plt.title('Branch vs Students')
plt.show()
```
