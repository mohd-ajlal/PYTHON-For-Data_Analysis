```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris['Species'] = iris['Species'].replace(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], [0, 1, 2])
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'], c=iris['Species'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
plt.scatter(iris['SepalLengthCm'], iris['SepalWidthCm'], c=iris['Species'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.colorbar(label='Species')
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap = 'jet', alpha = 0.7)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.colorbar(label='Species')
plt.show()
```
