```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'total_bill': [16.99, 10.34, 21.01, 23.68, 24.59],
    'tips': [1.01, 1.66, 3.50, 3.31, 3.61],
    'gender': ['Female', 'Male', 'Male', 'Male', 'Female']
}

df = pd.DataFrame(data)
male_d = df[df['gender'] == 'Male']
female_d = df[df['gender'] == 'Female']
plt.scatter(male_d['total_bill'], male_d['tips'], color='blue', label='Male')
plt.scatter(female_d['total_bill'], female_d['tips'], color='red', label='Female')
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.title('Tips vs Total Bill by Gender')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'total_bill': [16.99, 10.34, 21.01, 23.68, 24.59, 25.29, 8.77, 26.88],
    'tips': [1.01, 1.66, 3.50, 3.31, 3.61, 4.71, 2.00, 3.12],
    'sex': ['Female', 'Male', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female']
}

df = pd.DataFrame(data)

male_data = df[df['sex'] == 'Male']
female_data = df[df['sex'] == 'Female']

plt.figure(figsize=(10, 6))
plt.scatter(male_data['total_bill'], male_data['tips'], color='blue', marker='*', label='Male', s=100)
plt.scatter(female_data['total_bill'], female_data['tips'], color='red', marker='s', label='Female', s=100)

plt.xlabel('Total Bill ($)')
plt.ylabel('Tips ($)')
plt.title('Tips vs Total Bill by Gender')
plt.legend()

plt.show()
```
