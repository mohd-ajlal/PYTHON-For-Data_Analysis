```python
# for encryption purpose a web browser company wants to generate random values which have mean value equal to 0 and varience equal to 1. They want 5 randomly generated number in  2 different trials

# trial 1
trial1 = np.random.randn(5,3)
print(trial1.mean())
print(trial1.var())

# trial 2
trial2 = np.random.randn(5,3)
print(trial2.mean())
print(trial2.var())
```
