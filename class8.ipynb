```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv('tips.csv')
```

```python
plt.scatter(tips['total_bill'], tips['tip'],marker='o',  s = tips['tip']*20)
```

```python
plt.plot(tips['total_bill'], tips['tip'], 'r-', label='tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.legend()
plt.show()
```

```python
plt.plot(tips['total_bill'], tips['tip'],marker='o')
```

```python
plt.plot(tips['total_bill'], tips['tip'],marker='o', linestyle='none')
```

```python
plt.figure(figsize=(4, 6))
plt.plot(tips['total_bill'], tips['tip'],marker='o', linestyle='none')
```
