```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as pl
```

DATA

```python
# data - catagorical and numerical data

# catagorical data
# 1. Nominal data - no order
# 2. Ordinal data - order

# numerical data
# 1. Discrete data - countable
# 2. Continuous data - measurable

# data - structured and unstructured data
# structured data - data that can be stored in a table
# unstructured data - data that cannot be stored in a table

# data - primary and secondary data
# primary data - data collected by the researcher
# secondary data - data collected by someone else

# data - qualitative and quantitative data
# qualitative data - data that can be observed but not measured
# quantitative data - data that can be measured

# data - cross-sectional and time series data
# cross-sectional data - data collected at a single point in time
# time series data - data collected over a period of time

# data - raw and processed data
# raw data - data collected in its original form
# processed data - data that has been cleaned and organized
```

```python
price = [132, 220, 320, 402, 530]
quantity = [1, 2, 3, 4, 5]

pl.plot(price, quantity)
pl.show()
```

```python
data = pd.read_csv('./sharma-kohli.csv')
pl.plot(data['index'], data['V Kohli'], label='V Kohli', color='#cdb4db', linestyle='dashed', linewidth=3, marker='o', markerfacecolor='red', markersize=9)
pl.plot(data['index'], data['RG Sharma'], label='RG Sharma', color='#A2D2FF', linestyle='dashdot', linewidth=3, marker='o', markerfacecolor='blue', markersize=9)
pl.legend()
pl.xlabel('Year')
pl.ylabel('Runs')
pl.title('V Kohli vs RG Sharma')
pl.show()
```

```python
data = pd.read_csv('./sharma-kohli.csv')
pl.plot(data['index'], data['V Kohli'], label='V Kohli', color='#cdb4db', linestyle='dashed', linewidth=3, marker='o', markerfacecolor='red', markersize=9)
pl.plot(data['index'], data['RG Sharma'], label='RG Sharma', color='#A2D2FF', linestyle='dashdot', linewidth=3, marker='o', markerfacecolor='blue', markersize=9)
pl.legend()
pl.xlabel('Year')
pl.ylabel('Runs')
pl.xlim(2008, 2012)
pl.grid()
pl.title('V Kohli vs RG Sharma')
pl.show()
```

```python
data = pd.read_csv('./sharma-kohli.csv')
pl.scatter(data['index'], data['V Kohli'], label='V Kohli', color='#cdb4db', marker='o', s=100)
pl.scatter(data['index'], data['RG Sharma'], label='RG Sharma', color='#A2D2FF', marker='o', s=100)
pl.legend()
pl.xlabel('Year')
pl.ylabel('Runs')
# pl.xlim(2008, 2012)
pl.grid()
pl.title('V Kohli vs RG Sharma')
pl.show()
```
